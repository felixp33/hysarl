```python
import matplotlib.pyplot as plt
import numpy as np


def plot_data_series_with_average(data_series_list):
    # Create a figure and axis
    fig, ax = plt.subplots()

    # Plot each data series with lower opacity and same color
    for series in data_series_list:
        ax.plot(series, color='blue', alpha=0.15)

    # Calculate the average by time step
    average_series = np.mean(data_series_list, axis=0)

    # Plot the average series with higher opacity and same color
    ax.plot(average_series, color='blue', linewidth=3, label='Average')

    # Add legend
    ax.legend()

    # Show the plot
    plt.show()


# Example usage
data_series_list = [np.random.rand(100) for _ in range(10)]
plot_data_series_with_average(data_series_list)
```
